```python
# Part 1
## 1.1
df_customer = spark.read.csv('/FileStore/tables/customer.csv', header=True, inferSchema =True)
df_order = spark.read.csv('/FileStore/tables/order.csv', header=True, inferSchema =True)
df_customer.createOrReplaceTempView('Customer_table')
df_order.createOrReplaceTempView('Order_table')
```

```python
## 1.2
spark.sql("select FIRSTNAME as FirstName, LASTNAME as LastName, CITY as City, COUNTRY as Country from Customer_table order by COUNTRY, CITY").show()
```

```python
## 1.3
spark.sql("select distinct(CITY) as City from Customer_table where COUNTRY =\"AU\"").show()
```

```python
## 1.4
spark.sql("select COUNTRY as Country, count(CUST_ID) as Number from Customer_table group by COUNTRY HAVING Number > 10").show()
```

```python
## 1.5
spark.sql("select count(COUNTRY) as count from Customer_table where COUNTRY != \"CA\"").show()
```

```python
## 1.6
spark.sql("select FIRSTNAME as FirstName, LASTNAME as LastName from Customer_table JOIN Order_table on Customer_table.CUST_ID = Order_table.CUST_ID where TOTALAMOUNT > 15000").show()
```

```python
## 1.7
spark.sql("select ORDERNUMBER as OrderNumber, TOTALAMOUNT as TotalAmount, FIRSTNAME as FirstName, LASTNAME as LastName, CITY as City, COUNTRY as City from Customer_table JOIN Order_table on Customer_table.CUST_ID = Order_table.CUST_ID").show()
```
